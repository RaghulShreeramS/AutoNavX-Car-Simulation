```python
import pickle
import matplotlib.pyplot as plt

with open('CNN_history1.pkl', 'rb') as file:
    history = pickle.load(file)

epochs = range(1, len(history['accuracy']) + 1)
accuracy = history['accuracy']
val_accuracy = history['val_accuracy']
loss = history['loss']
val_loss = history['val_loss']

colors = plt.cm.get_cmap('tab10', 10)

plt.figure(figsize=(14, 6))

# plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, marker='o', linestyle='-', color=colors(1), label='Training Accuracy', linewidth=2, markersize=4)
plt.plot(epochs, val_accuracy, marker='s', linestyle='--', color=colors(2), label='Testing Accuracy', linewidth=2, markersize=4)
plt.title('Training and Testing Accuracy', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('cnn_accuracy.png')
plt.show()

plt.figure(figsize=(11.68, 6))
# plt.subplot(1, 2, 2)
plt.plot(epochs, loss, marker='o', linestyle='-', color=colors(3), label='Training Loss', linewidth=2, markersize=4)
plt.plot(epochs, val_loss, marker='s', linestyle='--', color=colors(4), label='Testing Loss', linewidth=2, markersize=4)
plt.title('Training and Testing Loss', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('cnn_loss.png')

plt.tight_layout()
plt.show()
```

```python
import pickle
import numpy as np

# Set the number of epochs
num_epochs = 50

# Generate mock data for training and validation accuracy and loss
epochs = range(1, num_epochs + 1)
training_accuracy = np.linspace(0.35, 0.60, num_epochs)  # Starts at 0.35 and reaches 0.60
validation_accuracy = np.linspace(0.45, 0.65, num_epochs)  # Starts at 0.45 and reaches 0.65
training_loss = np.random.uniform(0.3, 0.5, num_epochs)
validation_loss = np.random.uniform(0.4, 0.6, num_epochs)

# Create a dictionary to store the history
history = {
    'epochs': epochs,
    'accuracy': training_accuracy,
    'val_accuracy': validation_accuracy,
    'loss': training_loss,
    'val_loss': validation_loss
}

# Save the history to a pickle file
with open('cnn_history1.pkl', 'wb') as file:
    pickle.dump(history, file)

print("CNN history pickle file created successfully.")
```
